```python
import re
#separating age and gender
def extract_gender_age(value):
    match = re.match(r'(Women|Men)\s(\d{1,2}-\d{1,2})', value)
    if match:
        return match.groups()
    elif value == 'All Ages':
        return ('All', 'All')  # Handle 'All Ages'
    else:
        return ('Unknown', 'Unknown')  # Handle any unexpected values

# Apply the function and create Gender and Age columns
data[['Gender', 'Age']] = data['Target_Audience'].apply(lambda x: pd.Series(extract_gender_age(x)))
```
