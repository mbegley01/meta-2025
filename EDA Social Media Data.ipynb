```python

# Grouping by 'Campaign_Goal' and 'Channel_Used' and calculating the mean ROI
channel_goal_data = data.groupby(['Campaign_Goal', 'Channel_Used'])[['ROI']].mean()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Group by 'Campaign_Goal' and 'Channel_Used' and calculate the mean ROI
channel_goal_data = data.groupby(['Campaign_Goal', 'Channel_Used'])[['ROI']].mean()

# Pivot the table so that 'Campaign_Goal' is on the rows, 'Channel_Used' on the columns, and 'ROI' as the value
heatmap_data = channel_goal_data.pivot(index='Campaign_Goal', columns='Channel_Used', values='ROI')

# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("ROI by Campaign Goal and Channel Used")

# Create the heatmap
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")

# Add labels for the axes
plt.xlabel("Channel Used")
plt.ylabel("Campaign Goal")

# Display the plot
plt.show()
```
